### Preprocess CSV files
- 기존 CSV 파일 내 빈칸 채워넣기
- 식품 카테고리 확정

```python
import pandas as pd
```

1. (1) Food Data -> processed_food
- `제조국가` : "대한민국"으로 입력
- `유통바코드` : 전체 임의의 숫자로 입력
- `식품대분류` 및 `식품상세분류` : 식품 카테고리 추출

```python
food_data = pd.read_csv('original/processed_food.csv', header=1, low_memory=False)
food_data.head()
```

```python
id_col = list(range(1, len(food_data)+1))
food_data['id'] = pd.DataFrame({'id' : id_col})
```

```python
# is_blank = (processed_food['country_origin'] == '')
# processed_food.loc[is_blank, 'country_origin'] = '대한민국

food_data['country_origin'] = food_data['country_origin'].replace(to_replace=[None], value='대한민국')
food_data.head()
```

```python
import random

random.shuffle(id_col)
barcode_col = [i+8809647000000 for i in id_col]
food_data['barcode_no'] = pd.DataFrame({'barcode_no' : barcode_col})

food_data.head()
```

```python
processed_category_main = food_data['category_main'].to_list()
processed_category_sub = food_data['category_sub'].to_list()

processed_category_all = list(zip(processed_category_main, processed_category_sub))
print(processed_category_all[:10])
```

```python
food_data.to_csv('new/processed_food.csv', index=False)
```

2. Nutrition Data -> processed_nutrition, fresh_nutrition, cooked_nutrition
- 각각의 칸에 대해서 빈칸 또는 '-'를 모두 0으로 변경

```python
import os

def preprocess_nutrition(file_name):
  nutrition_data = pd.read_csv(os.path.join('original', file_name), header=1, low_memory=False)
  print('===== Original =====\n', nutrition_data.head())

  id_col = list(range(1, len(nutrition_data)+1))
  nutrition_data['id'] = pd.DataFrame({'id' : id_col})

  id_col_name = '_'.join([file_name.split('_')[0], 'food', 'id'])
  nutrition_data.rename(columns={'id' : id_col_name}, inplace=True)

  nutrition_data.replace(to_replace=[None, '-', 'Tr', '(Tr)'], value=0, inplace=True)
  nutrition_data.replace(to_replace=['미만$', '이하$', '음료$', '\)$', '^<'], value=0, inplace=True, regex=True)
  print('===== New =====\n', nutrition_data.head())

  nutrition_data.to_csv(os.path.join('new', file_name), index=False)
```

```python
preprocess_nutrition('processed_nutrition.csv')
```

```python
preprocess_nutrition('fresh_nutrition.csv')
```

```python
preprocess_nutrition('cooked_nutrition.csv')
```
